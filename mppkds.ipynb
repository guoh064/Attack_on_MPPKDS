$(\log_2 q, x, \log_2 p, n, \lambda, m) = (32,32,64,6,3,2)$

```python
x = 32; n = 6; l = 3; m = 2;
```

```python
while True:
    q = random_prime(2^(x+1), True, 2^x)
    if is_prime(2^x * q+1):
        p = 2^x*q+1
        break
q, p
```

```python
R = IntegerModRing(p-1); R
```

```python
l1 = 2; l2 = 1;
```

```python
beta = [[[R.random_element() for i in range(n+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]; beta
```

```python
f = [R.random_element() for i in range(l+1)]
h = [R.random_element() for i in range(l+1)]
f, h
```

```python
phi = [[convolution(f, beta[j1][j2]) for j2 in range(l2+1)] for j1 in range(l1+1)]
phi
```

```python
psi = [[convolution(h, beta[j1][j2]) for j2 in range(l2+1)] for j1 in range(l1+1)]
psi
```

```python
E_phi = [R.random_element() if i != 0 and i != n+l else R.zero() for i in range(n+l+1)] 
E_psi = [R.random_element() if i != 0 and i != n+l else R.zero() for i in range(n+l+1)]
E_phi, E_psi
```

```python
R_0 = R.random_element()
R_n = R.random_element()
R_0, R_n
```

```python
N_0 = [[[R.zero() if i != 0 else R_0 * beta[j1][j2][0] for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]; N_0
```

```python
N_n = [[[R.zero() if i != n+l else R_n * beta[j1][j2][n] for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]; N_n
```

```python
Phi = [[[R.zero() if i == 0 or i == n+l else phi[j1][j2][i] for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]; Phi
```

```python
Psi = [[[R.zero() if i == 0 or i == n+l else psi[j1][j2][i] for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]; Psi
```

```python
P = [[[R_0 * Phi[j1][j2][i] if (j1 != 0 or j2 != 0)else R_0 * (Phi[j1][j2][i] - E_phi[i]) for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]
P
```

```python
Q = [[[R_n * Psi[j1][j2][i] if (j1 != 0 or j2 != 0) else R_n * (Psi[j1][j2][i] - E_psi[i]) for i in range(n+l+1)] for j2 in range(l2+1)] for j1 in range(l1+1)]
Q
```

Below we give a forge.

```python
Mx = Matrix(R, (l1+1)*(l2+1), 4, [
    [S(Q[j1][j2])(mu),
     -S(P[j1][j2])(mu),
     -S(N_0[j1][j2])(mu),
     -S(N_n[j1][j2])(mu)] for j1 in range(l1+1) for j2 in range(l2+1)
])
Mx
```

```python
vq = vector([GF(q).zero() if _ != 0 else GF(q).one() for _ in range((l1+1)*(l2+1))]); vq
```

```python
Mq = (Mx % q).change_ring(GF(q)); Mq
```

```python
ansq = Mq \ vq; ansq
```

```python
R2 = IntegerModRing(2^x);
```

```python
M2 = (Mx % 2^x).change_ring(R2); M2
```

```python
for i in range(x):
    v2 = vector([R2.one()*2^i] + [R2.zero() for _ in range((l1+1)*(l2+1)-1)])
    try:
        ans2 = M2 \ v2
    except ValueError:
        continue
    else:
        break
ans2, v2
```

```python
_, m1, m2 = xgcd(2^x, q); m1, m2
```

```python
aa, bb, cc, dd = (ans2.lift() * m2 * q + ansq.lift() * m1 * 2^x) * R.one(); aa,bb,cc,dd
```

```python
ee = (v2[0].lift() * m2 * q + vq[0].lift() * m1 * 2^x) * R.one(); ee
```
